## Plotagem de campos vetoriais

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Setas (_quivers_)

```python
x1,x2 = np.meshgrid(np.linspace(-5,5,20),
                    np.linspace(-5,5,20))
```

```python
u1 = x1 + x2
u2 = 4*x1 + x2
```

```python
norma = np.sqrt(u1**2+u2**2)
```

```python
plt.quiver(x1, x2, u1, u2)
plt.title('Campo vetorial 2D')
plt.grid()
plt.show()
```

```python
plt.quiver(x1, x2, u1/norma, u2/norma)
plt.grid()
plt.show()
```

```python
norma = np.sqrt(u1**2+u2**2)

plt.quiver(x1, x2, 
           u1/norma, u2/norma,
           norma)

plt.colorbar()
plt.grid()
plt.show()
```

## Linhas de fluxo (_streamlines_)
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.streamplot.html

```python
plt.streamplot(x1, x2, u1, u2) 
plt.grid()
plt.show()
```

### Customizando o plot

```python
plt.streamplot(x1, x2, u1, u2,
               color = norma, 
               linewidth = 0.5, 
               cmap = 'jet',
               density = 2, 
               arrowstyle = '->', 
               arrowsize = 1.5) 
plt.grid()
plt.show()
```

## Camplo vetorial 3D

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')

x,y,z = np.meshgrid(np.arange(-1, 1, 0.25),
                    np.arange(-1, 1, 0.25),
                    np.arange(-1, 1, 0.25))

u = np.sin(x) * np.cos(y*z)
v = -y*z
w = np.cos(x) * np.sin(z)

ax.quiver(x, y, z, u, v, w, 
          length=0.2, linewidths= 0.7,
          color = 'black')

plt.show()
```
